```python
import urllib
import csv
import pandas as pd
from bs4 import BeautifulSoup
```

```python
df = pd.read_csv('AA_bestpictures.csv')
names = df.MovieName
```

```python
movie_name = []
critic_score = []
tomatometer_avg_score = []
audience_score = []
audience_avg_score = []
```

```python
length = len(names)
for x in range(0, length): 
    if x == 10 or x == 14 or x == 19 or x == 34 or x == 40 or x == 47 or x == 49 or x == 51 or x == 65 or x == 66 or x == 97 or x == 232 or x == 330 or x == 513:
        movie_name.append(names[x])
        critic_score.append("No info")
        tomatometer_avg_score.append("No info")
        audience_score.append("No info")
        audience_avg_score.append("No info")
    else:
    
        movie_name.append(names[x])
        print(names[x]+" good")
        current = str(names[x])
        current = current.strip()
        current = current.replace(' ', '%20')

        url="https://www.rottentomatoes.com/search/?search=" + current #url of interest
        req = urllib.request.Request(url, headers={'User-Agent' : "Magic Browser"}) 
        page = urllib.request.urlopen( req )
        soup = BeautifulSoup(page.read(), 'lxml') #parse page using xml parser
        movies = soup.find("li", { "class" : "bottom_divider clearfix" })
        section = movies.find('div', attrs={'class': 'poster'})
        link = section.findAll('a')
        link = str(link)
        link = link.split("\"")
        newurl = link[1]
        url="https://www.rottentomatoes.com" + newurl #url of interest
        req = urllib.request.Request(url, headers={'User-Agent' : "Magic Browser"}) 
        page = urllib.request.urlopen( req )
        soup = BeautifulSoup(page.read(), 'lxml') #parse page using xml parser

        movie_info = soup.find("div", {"id" : "scorePanel"})
        if movie_info is not None:
            tomatometer_avg = movie_info.find("div", {"class" : "superPageFontColor"})
            if tomatometer_avg is not None:
                tomatometer_avg = str(tomatometer_avg)
                tomatometer_avg = tomatometer_avg.split("\n")
                if (len(tomatometer_avg) < 3):
                    print (x)
                    tomatometer_avg_score.append("No info")
                else:
                    tomatometer_avg = tomatometer_avg[2]
                    tomatometer_avg = tomatometer_avg.strip()
                    tomatometer_avg_score.append(tomatometer_avg)
            else:
                 tomatometer_avg_score.append("No info")
        else:
            print(names(x))
            tomatometer_avg_score.append("No info")

        movie_info = soup.find("div", {"class" : "critic-score meter"})
        if movie_info is not None: 
            tomatometer_percent = movie_info.find("span", {"class" : "superPageFontColor"})
            tomatometer_percent = tomatometer_percent.find("span")
            tomatometer_percent = str(tomatometer_percent)
            tomatometer_percent = tomatometer_percent.split(">")
            tomatometer_percent = str(tomatometer_percent[1])
            tomatometer_percent = tomatometer_percent.split("<")
            tomatometer_percent = tomatometer_percent[0]
            critic_score.append(tomatometer_percent)
        else:
            print(names[x])
            critic_score.append("No info")

        movie_info = soup.find("div", {"class" : "audiencepanel"})
        if movie_info is not None:
            audience_avg = movie_info.find("div", {"class" : "audience-info"})
            audience_avg = str(audience_avg)
            audience_avg = audience_avg.split("\n")
            if (len(audience_avg) < 4):
                print (x)
                audience_avg_score.append("No info")
            else:
                audience_avg = audience_avg[3]
                audience_avg = audience_avg.strip()
                audience_avg_score.append(audience_avg)
        else:
            print(names[x])
            audience_avg_score.append("No info")

        movie_info = soup.find("div", {"class" : "audience-score"})
        if movie_info is not None:
            audience_percent = movie_info.find("span", {"class" : "superPageFontColor"})
            audience_percent = str(audience_percent)
            audience_percent = audience_percent.split(">")
            if (len(audience_percent) < 2):
                print(x)
                audience_score.append("No info")
            else:
                audience_percent = audience_percent[1]
                audience_percent = str(audience_percent)
                audience_percent = audience_percent.split("<")
                audience_percent = audience_percent[0]
                audience_percent = str(audience_percent)
                audience_percent = audience_percent.split("%")
                audience_percent = audience_percent[0]
                audience_score.append(audience_percent)
        else:
            print(names[x])
            audience_score.append("No info")
```

```python
print (len(movie_name))
print (len(critic_score))
print(len(tomatometer_avg_score))
data = pd.DataFrame({'Movie Name': movie_name, 'Tomatometer': critic_score, 'Average Critic Score': tomatometer_avg_score, 'Percent Audience Liked': audience_score, 'Average Audience Score': audience_avg_score}) #format into my DF
data = data[['Movie Name','Tomatometer', 'Average Critic Score', 'Percent Audience Liked', 'Average Audience Score']]
```
